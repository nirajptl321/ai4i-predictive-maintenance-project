# 05 Demo Usage

This notebook shows how to run the local demo. The demo loads the saved model and prints predictions for sample rows.

```python
from pathlib import Path
import subprocess

root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
result = subprocess.run(
    ["python", "demo/demo.py"],
    cwd=root,
    text=True,
    capture_output=True,
    check=True,
)
print(result.stdout)
```

The demo is local only and is not deployment software.
